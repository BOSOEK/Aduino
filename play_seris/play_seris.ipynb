### 판다스로 매핑시켜서 인코딩하기

```python
train_sec = pd.read_csv('./train.csv')
train_sec.head()
```

```python
map_sex = {'male' : 0, 'female' : 1}
col = ['Sex']
train_sec[col] = train_sec[col].applymap(map_sex.get)

map_sex = {'S' : 0, 'C' : 1, 'Q' : 2}
col = ['Embarked']
train_sec[col] = train_sec[col].applymap(map_sex.get)
# 텍스트 형을 숫자 형으로 인코딩
```

```python
train_sec.head()
```

> 이러면 행의 크기에 상관없이 인코딩 가능하다

### 'DataFrame.loc'으로 사용할 컬럼만 추출(이름 & PassengerId 제거)

```python
train_data = train_sec.loc[:, ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin', 'Embarked']]
train_label = train_sec.loc[:, ['Survived']]
```

```python
train_data.head()
```

### 데이터에서 NaN을 0으로 대체하기

```python
train_data = train_data.fillna(0) 
# fillna함수는 주어진 값으로 NaN을 대체함
```

```python
train_data.head()
```

### Cabin은 앞부분이 A, B, C, D, 0으로 시작한다. 그점을 이용해서 인코딩

> A, B, C, D로 시작하는 텍스트를 찾고 변환한다.

```python
#map_sex = {'A_' : 1, 'B_' : 2, 'C_' : 3, 'D_' : 4, 0:0}
#col = ['Cabin']
#train_data[col] = train_data[col].applymap(map_sex.get)
```

> 하지만 단순히 숫자로만 바꿔도 이 숫자는 아무런 의미가 없다.

### 텍스트의 의미를 제대로 담기 위해서 희소 벡터로 만들기 위해 원-핫 벡터(원-핫 인코딩)로 만든다.

```python
def one_hot_Encoding(data, column) :
    freq = data[column].value_counts()
    vocabulary = freq.sort_values(ascending = False).index
    
    for word in vocabulary :
        new_column = column + '_' + str(word)
        data[new_column] = 0
    
    for word in vocabulary :
        target_index = data[data[column] == word].index
        new_column = column + '_' + str(word)
        data.loc[target_index, new_column] = 1
    del(data[column])
```

### 특수한 텍스트로 시작하는 값을 바꾸는 코드 -> 못찾음 열 삭제

```python
train_data = train_data.loc[:, ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
```

```python
one_hot_Encoding(train_data, 'Pclass')
one_hot_Encoding(train_data, 'Sex')
one_hot_Encoding(train_data, 'Embarked')
```

```python
train_data.head()
```

```python
train_data.head()
```

### 비어있는 결측치 행을 제거한다

```python
train_data.dropna(inplace=True)
```

```python
train_data.head()
```

### 데이터를 7:3으로 분리한다

```python
X_train = train_data[:70000].to_numpy()
y_train = train_label[:70000].to_numpy()

X_test = train_data[70000:].to_numpy()
y_test = train_label[70000:].to_numpy()
```

```python
X_train.shape
```

# 표준화
> 상대적으로 값이 큰 Age, fare를 스케일 변환시킨다.

```python
age_min = np.min(X_test[:,2])
age_max = np.max(X_test[:,2])

Fare_min = np.min(X_test[:,5])
Fare_max = np.max(X_test[:,5])

X_train[:,2] = (X_train[:,2] - age_min)/(age_max - age_min)
X_test[:,2] = (X_test[:,2] - age_min) / (age_max - age_min)

X_train[:,5] = (X_train[:,5] - Fare_min) / (Fare_max - Fare_min)
X_test[:,5] = (X_test[:,5] - Fare_min) / (Fare_max - Fare_min)
```

```python
X_train
```

### 모델 생성 & 학습

```python
model = keras.Sequential()
model.add(Dense(128, activation = 'relu'))
model.add(Dropout(0.10))
model.add(Dense(64, activation = 'relu'))
model.add(Dropout(0.10))
model.add(Dense(32, activation = 'relu'))
model.add(Dropout(0.10))
model.add(Dense(16, activation = 'relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.005),
             loss='binary_crossentropy',
             metrics=['binary_accuracy'])
```

```python
model.fit(X_train, y_train, epochs=50)
```

```python
test_loss, test_acc = model.evaluate(X_test,  y_test, verbose=2)
```

```python
map_sex = {'male' : 0, 'female' : 1}
col = ['Sex']
test[col] = test[col].applymap(map_sex.get)

map_sex = {'S' : 0, 'C' : 1, 'Q' : 2}
col = ['Embarked']
test[col] = test[col].applymap(map_sex.get)

test = test.loc[:, ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin', 'Embarked']]

test = test.fillna(0) 

test = test.loc[:, ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]

one_hot_Encoding(test, 'Pclass')
one_hot_Encoding(test, 'Sex')
one_hot_Encoding(test, 'Embarked')

test.dropna(inplace=True)

predictions = model.predict(test)
```
